#### Bài 1: Nhập vào 2 cạnh của một hình chữ nhật. In ra màn hình diệntíchvàchu vi của nó.

```python
def tinh_hcn (a,b):
    p = 2*(a+b)
    s = a*b
    return p, s
a, b = map(float,input().split())
p, s = tinh_hcn(a,b)
print(f"Chu vi: {p}")
print(f"dien tich: {s}")
```

Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểmtra xem nó có thể làđộ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra mànhình

```python
import math
def check_tam_giac(a, b ,c):
    if a + b > c and b + c > a and c + a > b:
        return True
    return False

a, b, c = map(float,input().split())
tam_giac_can = (a == b or b == c or c == a)
tam_giac_deu = (a == b == c)
tam_giac_vuong = (a**2 + b**2 == c**2) or (b**2 + c**2 == a**2) or (a**2 + c**2 == b**2)
p = a+b+c
s = math.sqrt(p*(p-a)*(p-b)*(p-c))

if check_tam_giac(a, b, c):
    if tam_giac_deu:
        print ("Tam giác đều")
    elif tam_giac_can and tam_giac_vuong:
        print ("Tam giác vuông cân")
    elif tam_giac_vuong:
        print ("Tam giác vuông")
    elif tam_giac_can:
        print ("Tam giác cân")
    else:
        print ("Tam giác thường")
else:
    print ("KO phai tam giac")                    
```

Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
import math
def giai_pt_bac_2 (a, b, c):
    delta = b ** 2 - 4*a*c
    if delta > 0:
        nghiem1 = (b + math.sqrt(delta))/2*a
        nghiem2 = (b - math.sqrt(delta))/2*a
        return nghiem1, nghiem2

a, b, c = map(int,input().split())
nghiem1, nghiem2 = giai_pt_bac_2(a, b, c)
print (nghiem1, nghiem2)
```

### Bài 8: Giải hệ phương trình tuyến tính:
#### ax + by = m
#### cx + dy = n

```python
a, b, m = map(float, input().split())
c, d, n = map(float, input().split())

D = a * d - b * c
Dx = m * d - b * n
Dy = a * n - m * c

if D != 0:
    x = Dx / D
    y = Dy / D
    print(f"x = {x:.2f}, y = {y:.2f}")
elif Dx == 0 and Dy == 0:
    print("He vo so nghiem")
else:
    print("He vo nghiem")
```

### Bài 9: Nhập vào thời gian 1 công việc nào đó là x giây. Hãy chuyển đổi và viết ra màn hình số thời gian trên dưới dạng bằng bao nhiêu giờ, bao nhiêu phút, bao nhiêu giây
