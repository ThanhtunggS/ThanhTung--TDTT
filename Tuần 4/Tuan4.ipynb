### W4A1 
[SumOfNumBers] 
Viết chương trình Python sử dụng một trong 2 loại vòng lặp for hoặc while để 
tính tống các số nguyên từ 1 đến n. (n <= 1000) 

```python
def sum_of_number(n):
    tong = 0
    for i in range (1, n+1):
        tong += i
    return tong
while (True):
    try:
        n = int(input("Nhập số nguyên n: "))
        if n <= 1000:
            tong = sum_of_number(n)
            print (tong)
            break
        else:
            print ("n lớn hơn 1000")
            continue
    except ValueError:
        print ("Vui lòng nhập một số nguyên hợp lệ!")   
        continue     
```

### W4A3 
[Factorial]  
Viết chương trình tính giai thừa của một số nguyên n cho trước (n!=1×2×...×n) 
với 0 < n < 100

```python
def tinh_giai_thua(n):
    ket_qua = 1
    if n<0:
        return "Lỗi do n < 0"
    if n==0:
        ket_qua = 1
        return ket_qua
    else:    
        for i in range (1,n+1):
            ket_qua *= i
        return ket_qua
n = int(input("Type n: "))
n = tinh_giai_thua(n)
print (n)    
```

### W4A6 
"[SumPrimesInRange] 
Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn 
[a, b]. Mỗi dòng là một bộ test (a b)."

```python
so_nt = []
def check_snt(n):
    if n == 2: return True
    else: 
        for i in range (2, n**0.5 + 1):
            if n % i == 0: return False
        return True

a, b = map(int,input().split())    
for i in range (a, b):
    if check_snt(i):
        so_nt.append (i)

print (sum(so_nt))
```

#### W4A7 
"[LargestPrimeFactor] 
Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n." 


```python
import math
def kiem_tra_snt(n):
    if n < 2:
        return False
    if n == 2 or n == 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    for x in range(5, int(math.sqrt(n)) + 1, 6):
        if n % x == 0 or n % (x + 2) == 0:
            return False
    return True

def primes (n):
    danh_sach_snt = []
    for x in range (n+1):
        if kiem_tra_snt(x):
            danh_sach_snt.append(x)
    return danh_sach_snt

try:
    n = int(input("Type n: "))
    primes_list = primes(n)        
    if not primes_list:
        print ("KO có SNT") 
    else:    
        max_prime = max(primes_list)
        print (f"SNT lớn nhất là: {max_prime}")
except ValueError:
    print ("Hãy nhập giá trị số hợp lệ")



```

#### W4A8 
ReverseAddPalindrome
Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số 
đối xứng (palindrome). In ra số bước và giá trị palindrome thu được. 

```python
def palindrome (n):
    reverse = 0
    while n > 0:
        chu_so = n % 10
        reverse = reverse * 10 + chu_so
        n //= 10
    return reverse

n = int(input())

while (n != palindrome(n)):
    n += palindrome(n)
print (n)    
```

#### W4A10 
"[LongestCollatzUpToN] 
Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất 
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số 
nhỏ nhất."

```python


def find_biggest_collazt (n):
    if n == 1:
        print ("False")
    else:
        ham_dem = []
        for i in range (2,n+1):
            collatz_list = []
            while i > 1:
                if i % 2 == 0:
                    i /= 2
                    collatz_list.append(i) 
                    continue
                else:
                    i = 3 * i + 1
                    collatz_list.append(i)
                    continue   
            a = len(collatz_list)
            ham_dem.append(a)
        x = max(ham_dem)
        return x

n = int(input("Nhap n: ")) 
x = find_biggest_collazt(n)    
print (x)   
```

[ClosePairNumber] 
Hai số nguyên dương tạo thành một cặp số thân thiết khi chúng tuân theo quy luật: 
Số này bằng tổng tất cả các ước của số kia (trừ chính nó) và ngược lại. 
Ví dụ 220 và 284 là một cặp số thân thiết 
220 ngoài chính nó ra, nó còn có 11 ước số khác là: 1, 2, 4, 5, 10, 11, 20, 44, 55, 
110. Tổng của 11 ước này cho chúng ta 284 
284 ngoài bản thân nó, còn có 5 ước số khác là: 1, 2, 4, 71, 142. Tổng của chúng 
vừa đúng bằng 220 
Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không. 
Nếu có, in ra màn hình "true", ngược lại, in ra "false"

```python
def tong_uoc (n):
    uoc = []
    for i in range (1, n):
        if n % i == 0:
            uoc.append (i)
    return sum(uoc)

a, n = map(int,input().split())
if a == tong_uoc(n): print ("true")
else: print ("Flase")        
        
```

### W4A14 
[GCD] 
Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. Viết chương trình 
tìm ước chung lớn nhất của hai số m và n.

```python
def uoc_chung (m,n):
    uocc = []
    for i in range (1,m + 1):
        if m % i == 0 and n % i == 0:
            uocc.append(i)
    return max(uocc)        

m, n = map(int,input().split())
print (uoc_chung(m,n))
```

Từ bài toán dân gian trên, mở rộng thành chương trình nhận đầu vào là tổng số

```python
def check_gacho (con, chan):
    for i in range (1, con):
        j = con - i
        if i * 2 + j * 4 == chan: return True
    return False

con, chan = map(int,input().split())
if check_gacho(con, chan): print (con, chan)
else: print ("invalid")  
```

### W4A20 Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không

```python
def check_2k (n):
    for i in range (1, n//2):
        if 2 ** i == n:
            return True
    return False
n = int(input())
print (check_2k(n))
```
