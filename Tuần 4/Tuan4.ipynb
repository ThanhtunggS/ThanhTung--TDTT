### W4A1 
[SumOfNumBers] 
Viết chương trình Python sử dụng một trong 2 loại vòng lặp for hoặc while để 
tính tống các số nguyên từ 1 đến n. (n <= 1000) 

```python
def sum_of_number(n):
    tong = 0
    for i in range (1, n+1):
        tong += i
    return tong
while (True):
    try:
        n = int(input("Nhập số nguyên n: "))
        if n <= 1000:
            tong = sum_of_number(n)
            print (tong)
            break
        else:
            print ("n lớn hơn 1000")
            continue
    except ValueError:
        print ("Vui lòng nhập một số nguyên hợp lệ!")   
        continue     
```

### W4A3 
[Factorial]  
Viết chương trình tính giai thừa của một số nguyên n cho trước (n!=1×2×...×n) 
với 0 < n < 100

```python
def tinh_giai_thua(n):
    ket_qua = 1
    if n<0:
        return "Lỗi do n < 0"
    if n==0:
        ket_qua = 1
        return ket_qua
    else:    
        for i in range (1,n+1):
            ket_qua *= i
        return ket_qua
n = int(input("Type n: "))
n = tinh_giai_thua(n)
print (n)    
```

#### W4A7 
"[LargestPrimeFactor] 
Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n." 


```python
import math
def kiem_tra_snt(n):
    if n < 2:
        return False
    if n == 2 or n == 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    for x in range(5, int(math.sqrt(n)) + 1, 6):
        if n % x == 0 or n % (x + 2) == 0:
            return False
    return True

def primes (n):
    danh_sach_snt = []
    for x in range (n+1):
        if kiem_tra_snt(x):
            danh_sach_snt.append(x)
    return danh_sach_snt

try:
    n = int(input("Type n: "))
    primes_list = primes(n)        
    if not primes_list:
        print ("KO có SNT") 
    else:    
        max_prime = max(primes_list)
        print (f"SNT lớn nhất là: {max_prime}")
except ValueError:
    print ("Hãy nhập giá trị số hợp lệ")



```

#### W4A8 
ReverseAddPalindrome
Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số 
đối xứng (palindrome). In ra số bước và giá trị palindrome thu được. 

```python
def palindrome (n):
    reverse = 0
    while n > 0:
        chu_so = n % 10
        reverse = reverse * 10 + chu_so
        n //= 10
    return reverse

n = int(input())

while (n != palindrome(n)):
    n += palindrome(n)
print (n)    
```

#### W4A10 
"[LongestCollatzUpToN] 
Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất 
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số 
nhỏ nhất."

```python


def find_biggest_collazt (n):
    if n == 1:
        print ("False")
    else:
        ham_dem = []
        for i in range (2,n+1):
            collatz_list = []
            while i > 1:
                if i % 2 == 0:
                    i /= 2
                    collatz_list.append(i) 
                    continue
                else:
                    i = 3 * i + 1
                    collatz_list.append(i)
                    continue   
            a = len(collatz_list)
            ham_dem.append(a)
        x = max(ham_dem)
        return x

n = int(input("Nhap n: ")) 
x = find_biggest_collazt(n)    
print (x)   
```

### W4A14 
[GCD] 
Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. Viết chương trình 
tìm ước chung lớn nhất của hai số m và n.

```python
def uoc_chung (m,n):
    uocc = []
    for i in range (1,m + 1):
        if m % i == 0 and n % i == 0:
            uocc.append(i)
    return max(uocc)        

m, n = map(int,input().split())
print (uoc_chung(m,n))
```

#### W4A22 Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
def dem (n):
    chan = [int(digit) for digit in str(n) if int(digit) % 2 == 0] 
    le = [int(digit) for digit in str(n) if int(digit) % 2 != 0]
    return (set(chan), set(le)) 

n = int(input())
chan, le = dem(n)
print(f"chan: {', '.join(map(str, chan))}") #Dùng join để bỏ dấu {}
print(f"le: {', '.join(map(str, le))}")
```

 Cho S(k) = 1 + 2 + 3 + … + k,  
 Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n

```python
def tinh_k (n):
    for i in range (1,n):
        k = i*(i-1)/2
        if k >= n:
            i -= 1
            k = i*(i-1)/2
            return (k, i)

n = int(input())
k, i = tinh_k (n)
print (i)
print (f"n = {n} and k = {k:g}")  #Thêm :g sau {k} để loại bỏ phần .0 (nếu k là 10.5 thì ko loại bỏ .5, chỉ loại bỏ .0)  (print(*...)) để loại bỏ [] 
# Thuật toán chính: Tính k liên tục, đến khi k >= n thì giảm i xuống 1, từ đó k giảm theo (bé hơn n)  
```

#### W4A24
Nhập vào A, tìm n nhỏ nhất sao cho
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A

```python
def tim_n (A):
    i = 0
    k = 0
    while k <= A:
        i += 1
        k += 1/i
    return (k, i)

A = int(input())
k, i = tim_n(A)  
print (A)
print (f"n = {k:.2f}, i = {i}")  

#Thuật toán: tính 1/1 + 1/2 +... Khi tính đc một số lớn hơn A, máy quay lại vòng lặp while và thấy ko thỏa mãn k <= A thì sẽ ko cộng tiếp
# Như vậy là ta cộng trước đến khi nó lớn hơn A, máy check thấy lớn hơn rồi thì ko cộng nữa
```

Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ
bằng tổng hai số trước đó
. Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt
quá A

```python
def calculate_fibo (A):
    fibo = [1, 1]  #Lấy nguồn của dãy Fibo để sau cộng vào
    if A == 1:
        return 1
    elif A > 1:
        i = 3
        while True:
            k = int(sum(fibo[i - 3 : i - 1]))
            fibo.append (k)
            i += 1
            if k > A:
                k = int(fibo [-2]) #[-2] là để truy cập vị trí ngay trước vị trí cuối cùng
                del fibo [-1] #[-1] là để truy cập vị trí cuối cùng
                return (k, i, fibo)
            
A = int(input())
k, i, z = calculate_fibo(A)
print (f"A = {A}")
print (f"số Fibo = {k}")
print (f"Dãy Fibo: {z}")

#Thuật toán: Để tính dãy Fibo, ta lấy nguồn của nó là hai số [1, 1], rồi tính theo công thức 1 + 1 = 2, thêm 2 vào dãy fibo, rồi 1 + 2 = 3,
#Sau đó thêm 3 vào dãy Fibo và lặp lại quá trình
#Khi nào số fibo (k) lớn hơn A, thì ta lấy giá trị ở trước k, và sao đó xóa đi k trong dãy fibo
```

### W4A31
Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in
thường, bao nhiêu ký tự số

```python
def count_characters(s):
    uppercase_count = 0
    lowercase_count = 0
    digit_count = 0
    
    for char in s:
        if char.isupper():
            uppercase_count += 1
        elif char.islower():
            lowercase_count += 1
        elif char.isdigit():
            digit_count += 1
            
    return uppercase_count, lowercase_count, digit_count

n = input()
upper, lower, digits = count_characters(n)
print(f"Ký tự in hoa: {upper}")
print(f"Ký tự in thường: {lower}")
print(f"Ký tự số: {digits}")
```
