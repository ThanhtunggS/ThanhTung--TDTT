### Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. Người ra xây một khu vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. Nhập vào a, b. Hãy tính phần diện tích còn để trồng cây xung quanh là bao nhiêu. Biết pi = 3.14, lấy kết quả đến 2 chữ số phần thập phân.

```python
def tinh_dien_tich(a, b):
    s_square = a*b
    s_cirle = (b/2)*3.14
    s_garden = s_square - s_cirle
    return s_garden
a, b = map(int,input().split())
c = tinh_dien_tich(a, b)
print (c)
```

### Bài 6. Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không, nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”.

```python
import math
def kiem_tra_tam_giac (a, b, c):
    if a + b > c and b + c > a and c + a > b:
        return True
    else:
        return False
def tinh_dien_tich (a, b, c):
    p = (a+b+c)/2
    dien_tich = math.sqrt(p*(p-a)*(p-b)*(p-c))
    return dien_tich
a, b, c = map(int,input().split())
if kiem_tra_tam_giac(a, b, c):
    dien_tich = tinh_dien_tich(a,b,c)
    print(f"{dien_tich:.1f}")
else:
    print ("Khong phai 3 canh cua tam giac")    
```

### Bài 8: Nhập vào họ tên chủ hộ và chỉ số điện kế tháng trước và chỉ số điện kế tháng này, tính tiền điện tháng này cho hộ biết:
- Từ 0 – 50kWh: 1.984 đồng/kWh
- Bậc 2: Từ 51 – 100 kWh: 2.050 đồng/kWh
- Bậc 3: Từ 101 – 200 kWh: 2.380 đồng/kWh
- Bậc 4: Từ 201 – 300 kWh: 2.998 đồng/kWh
- Bậc 5: Từ 301 – 400 kWh: 3.350 đồng/kWh
- Bậc 6: Từ 401 trở đi kWh: 3.460 đồng/kWh
Tính thêm tiền VAT 8%. Chỉ lấy đến đơn vị đồng

```python
def tinh_tien_dien (a):
    if 0 <= a <= 50:
        tong_tien = round(1.984*a)
        return tong_tien
    elif 51 <= a <= 100:
        tong_tien = round(2.050*a)
        return tong_tien
    elif 101 <= a <= 200:
        tong_tien = round(2.380*a)  
        return tong_tien     
    elif 201 <= a <= 300:
        tong_tien = round(2.998*a)  
        return tong_tien  
    elif 301 <= a <= 400:
        tong_tien = round(3.350*a)  
        return tong_tien    
    elif 401 <= a:
        tong_tien = round(3.460*a)
        return tong_tien
a = int(input())
tien_dien = tinh_tien_dien(a)
print (tien_dien)    
```

### 6. Kiểm tra ký tự nhập vào là chữ hay số: Viết chương trình nhập một ký tự, kiểm tra xem ký tự đó là chữ cái hay số.

```python
ky_tu = input()
if ky_tu.isdigit():
        print(f"'{ky_tu}' là một chữ số.")
elif ky_tu.isalpha():
        print(f"'{ky_tu}' là một chữ cái.")
else:
        print(f"'{ky_tu}' không phải là chữ cái cũng không phải chữ số.")
```
