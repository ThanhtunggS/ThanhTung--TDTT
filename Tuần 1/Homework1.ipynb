### Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. Người ra xây một khu vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. Nhập vào a, b. Hãy tính phần diện tích còn để trồng cây xung quanh là bao nhiêu. Biết pi = 3.14, lấy kết quả đến 2 chữ số phần thập phân.

```python
def tinh_dien_tich(a, b):
    s_square = a*b
    s_cirle = (b/2)*3.14
    s_garden = s_square - s_cirle
    return s_garden
a, b = map(int,input().split())
c = tinh_dien_tich(a, b)
print (c)
```

### Viết chương trình nhập vào một “kí tự” c, nếu nó là chữ hoa thì in ra chữ thường tương ứng và ngược lại nếu nó là chữ thường thì in ra chữ hoa tương ứng của nó. (xem trước cấu trúc lựa chọn: if….else, ma)


```python
c = input()
ascii_val = ord(c)
if 65 <= ascii_val <= 90:
    print(chr(ascii_val + 32))
elif 97 <= ascii_val <= 122:
    print(chr(ascii_val - 32))
else:
    print(c)
```

### Bài 5: Nhập vào chữ cái hoa A, in ra chữ cái thường liền trước chữ cái a thường tương ứng của A. Xét TH khi c = ‘A’ là đặc biệt.

```python
def ham(c):
    a = ord(c)
    if 65 < a <= 90:
        return chr(a+31)
    elif a == 65:
        return "TH đặc biệt"
c = input()
d = ham(c)
print(d)
```
