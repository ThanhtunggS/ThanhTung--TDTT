#### W6A1
Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng
giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả.
Yêu cầu in ra đúng dạng list Python, ví dụ [1, 2, 3]

```python
a = list(map(int,input().split()))
result = []
for i in a:
    if i not in result:
        result.append(i)
final = sorted(result)        
print (a, final)        
```

```python
a = list(map(int,input().split()))
result = []
b = set()
for i in a:
    if i not in b:
        result.append(i)
        b.add(i)
print (a, result)        
```

#### W6A2 
Nhập một dòng số nguyên. In ra list tổng tích lũy (phần tử i là tổng từ đầu 
tới vị trí i), giữ nguyên thứ tự 

```python
def tinh_tong_tich_kuy (n):
    tap_so = [0, 1]
    for i in range (1, n):
        tong = sum(tap_so[:i+1])
        tap_so.append (tong)
    return tap_so
n = int(input())
print (*tinh_tong_tich_kuy(n), sep = ", ")
```
